```python
import os, json, random
import torch
import PIL.Image
import numpy as np
import pandas as pd
from tqdm import tqdm
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import torchvision.models as tv_models
from scipy.stats import spearmanr, pearsonr

DEVICE     = 'cuda' if torch.cuda.is_available() else 'cpu'
JSON_PATH  = '/content/TrainSet/train_SIQA-S.jsonl'
IMAGE_DIR  = '/content/TrainSet'
SEED       = 42
VAL_RATIO  = 0.2
EPOCHS     = 10        # more epochs since we are fine-tuning the full model
BATCH_SIZE = 32

# Differential learning rates
LR_BACKBONE = 1e-5    # 10x lower — avoids destroying ImageNet features
LR_NEW      = 1e-4    # faster for HyperNet + head

SCORE_MIN, SCORE_MAX = 1.0, 5.0

random.seed(SEED)
torch.manual_seed(SEED)
print(f'Device: {DEVICE}')
```

```python
def safe_open(path):
    """Returns PIL image or None if file is corrupted."""
    try:
        img = PIL.Image.open(path)
        img.verify()
        img = PIL.Image.open(path).convert('RGB')  # must re-open after verify
        return img
    except Exception:
        return None

def normalize(s):   return (s - SCORE_MIN) / (SCORE_MAX - SCORE_MIN)
def denormalize(s): return float(s) * (SCORE_MAX - SCORE_MIN) + SCORE_MIN

print('Helpers ready')
```

```python
with open(JSON_PATH) as f:
    raw_data = [json.loads(line) for line in f]
print(f'Total records: {len(raw_data)}')

# Pre-scan — removes corrupted images once so training never crashes mid-epoch
bad_paths  = set()
clean_data = []
for item in tqdm(raw_data, desc='Scanning for corrupted images'):
    path = os.path.join(IMAGE_DIR, item['image'])
    if safe_open(path) is None:
        bad_paths.add(path)
    else:
        clean_data.append(item)

print(f'Clean: {len(clean_data)} | Corrupted (removed): {len(bad_paths)}')
```

```python
random.shuffle(clean_data)
split      = int(len(clean_data) * (1 - VAL_RATIO))
train_data = clean_data[:split]
val_data   = clean_data[split:]

print(f'Train: {len(train_data)} | Val: {len(val_data)}')
print('Val set is held out — never seen during training')
```

```python
# Matches the original HyperIQA paper preprocessing
train_transform = T.Compose([
    T.Resize((512, 512)),
    T.RandomCrop(384),
    T.RandomHorizontalFlip(),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std =[0.229, 0.224, 0.225]),
])

val_transform = T.Compose([
    T.Resize((512, 512)),
    T.CenterCrop(384),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std =[0.229, 0.224, 0.225]),
])

print('Transforms ready')
```

```python
class SIQADataset(Dataset):
    def __init__(self, records, image_dir, transform):
        self.records   = records
        self.image_dir = image_dir
        self.transform = transform

    def __len__(self):
        return len(self.records)

    def __getitem__(self, idx):
        item  = self.records[idx]
        path  = os.path.join(self.image_dir, item['image'])
        image = PIL.Image.open(path).convert('RGB')
        image = self.transform(image)

        return {
            'image': image,
            'gt_p':  torch.tensor(normalize(float(item['perception_raing'])),
                                  dtype=torch.float),
            'gt_k':  torch.tensor(normalize(float(item['knowledge_rating'])),
                                  dtype=torch.float),
            'pid':   item['pid'],
            'path':  path,
            # Keep raw GT for metrics at inference
            'raw_p': float(item['perception_raing']),
            'raw_k': float(item['knowledge_rating']),
        }

train_dataset = SIQADataset(train_data, IMAGE_DIR, train_transform)
val_dataset   = SIQADataset(val_data,   IMAGE_DIR, val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                           shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE,
                           shuffle=False, num_workers=2, pin_memory=True)

print(f'Train: {len(train_dataset)} samples = {len(train_loader)} batches of {BATCH_SIZE}')
print(f'Val  : {len(val_dataset)} samples = {len(val_loader)} batches of {BATCH_SIZE}')
```

```python
class HyperIQA(nn.Module):
    """
    HyperIQA (Su et al., CVPR 2020) for SIQA-S two-output scoring.

    Backbone : ResNet50 pretrained on ImageNet — fully unfrozen during training
               to match the 'HyperIQA (Trained)' benchmark setup.
    HyperNet : Content-adaptive FC layers that generate quality-aware weights.
    Head     : Two outputs (perception, knowledge) with Sigmoid → [0,1].
    """
    def __init__(self, pretrained=True):
        super().__init__()
        weights       = tv_models.ResNet50_Weights.IMAGENET1K_V1 if pretrained else None
        resnet        = tv_models.resnet50(weights=weights)
        self.backbone = nn.Sequential(*list(resnet.children())[:-2])  # output (B,2048,H,W)
        self.pool     = nn.AdaptiveAvgPool2d(1)

        self.hyper = nn.Sequential(
            nn.Linear(2048, 1024), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(1024, 512),  nn.ReLU(), nn.Dropout(0.25),
            nn.Linear(512,  256),  nn.ReLU(),
        )
        self.head = nn.Sequential(
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 2),
            nn.Sigmoid(),
        )

    def forward(self, x):
        feat = self.pool(self.backbone(x)).flatten(1)  # (B, 2048)
        return self.head(self.hyper(feat))             # (B, 2)


model = HyperIQA(pretrained=True).to(DEVICE)

# ── Differential learning rates ───────────────────────────────────────────────
# Backbone gets 10x lower LR to preserve ImageNet features while still adapting.
# HyperNet + head get the full LR since they are newly initialised.
optimizer = optim.Adam([
    {'params': model.backbone.parameters(), 'lr': LR_BACKBONE},
    {'params': model.hyper.parameters(),    'lr': LR_NEW},
    {'params': model.head.parameters(),     'lr': LR_NEW},
])
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)
criterion = nn.MSELoss()

trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total     = sum(p.numel() for p in model.parameters())
print(f'Trainable: {trainable:,} / {total:,} parameters')
print(f'Backbone LR: {LR_BACKBONE} | HyperNet+Head LR: {LR_NEW}')
```

```python
print(f'Training for {EPOCHS} epochs on {len(train_dataset)} samples...')
print(f'Each epoch = {len(train_loader)} batches of {BATCH_SIZE} images')
print('(Full model fine-tune — backbone + HyperNet + head)\n')

for epoch in range(EPOCHS):
    model.train()
    total_loss, processed = 0.0, 0

    for batch in tqdm(train_loader, desc=f'Epoch {epoch+1}/{EPOCHS}'):
        images = batch['image'].to(DEVICE)
        labels = torch.stack(
            [batch['gt_p'], batch['gt_k']], dim=1
        ).to(DEVICE)                          # (B, 2) normalized to [0,1]

        optimizer.zero_grad()
        outputs = model(images)               # (B, 2)
        loss    = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)
        processed  += images.size(0)

    scheduler.step()
    avg = total_loss / processed

    # Convert loss back to [1,5] scale for interpretability
    # MSE on [0,1] → multiply by (5-1)^2=16 to get [1,5] scale MSE
    approx_mae = (avg * 16) ** 0.5
    print(f'  Epoch {epoch+1:2d}/{EPOCHS} | '
          f'Loss (normalized): {avg:.6f} | '
          f'Approx MAE [1-5]:  {approx_mae:.4f} | '
          f'LR backbone: {scheduler.get_last_lr()[0]:.2e}')

print('\nTraining complete.')
```

```python
# ── Evaluate on held-out val set only ────────────────────────────────────────
print(f'Evaluating on {len(val_dataset)} held-out validation samples...')
model.eval()
all_results = []

with torch.no_grad():
    for batch in tqdm(val_loader, desc='Inference'):
        images = batch['image'].to(DEVICE)
        outs   = model(images).cpu().numpy()   # (B, 2) in [0,1]

        for i in range(outs.shape[0]):
            pid = batch['pid'][i]
            pid = pid.item() if hasattr(pid, 'item') else pid
            all_results.append({
                'pid':    pid,
                'path':   batch['path'][i],
                # raw_p / raw_k come directly from the dataset, no recomputation needed
                'gt_p':   float(batch['raw_p'][i]),
                'gt_k':   float(batch['raw_k'][i]),
                'pred_p': denormalize(outs[i, 0]),
                'pred_k': denormalize(outs[i, 1]),
            })

print(f'Inference done — {len(all_results)} samples')
```

```python
df = pd.DataFrame(all_results)

s_p, _ = spearmanr(df['gt_p'], df['pred_p'])
p_p, _ = pearsonr( df['gt_p'], df['pred_p'])
s_k, _ = spearmanr(df['gt_k'], df['pred_k'])
p_k, _ = pearsonr( df['gt_k'], df['pred_k'])

score_p = max((s_p + p_p) / 2, 0) * 100
score_k = max((s_k + p_k) / 2, 0) * 100
final   = (score_p + score_k) / 2

print('=' * 60)
print(f'  Evaluated on  : {len(df)} held-out val samples')
print(f'  PERCEPTION    SRCC: {s_p:.4f} | PLCC: {p_p:.4f}  → {score_p:.2f}')
print(f'  KNOWLEDGE     SRCC: {s_k:.4f} | PLCC: {p_k:.4f}  → {score_k:.2f}')
print(f'  FINAL SCORE   : {final:.2f}')
print('=' * 60)
print()
print('Benchmark reference (from the challenge paper):')
print('  CLIP-IQA+ (Trained) : 74.1')
print('  HyperIQA  (Trained) : 83.5   ← target')
print('  InternVL3 (FT)      : 89.8')
```

```python
# --- IMPROVED INFERENCE CELL ---
import json
import os
import torch
import PIL.Image
from tqdm import tqdm

# --- Configuration ---
TEAM_NAME   = "YourTeamName"
MODEL_NAME  = "HyperIQA-Colab"
TEST_JSONL  = "/content/SIQA-test.jsonl"
OUTPUT_JSON = "/content/submission_S.json"

def safe_predict_load(path):
    """Attempt to open and verify image; returns PIL image or None."""
    try:
        with open(path, 'rb') as f:
            img = PIL.Image.open(f)
            img.load() # Force loading to catch truncation errors early
            return img.convert('RGB')
    except Exception:
        return None

print(f"Generating predictions for {TEST_JSONL}...")
model.eval()
predictions = []

if not os.path.exists(TEST_JSONL):
    print(f"⚠️ Error: {TEST_JSONL} not found!")
else:
    with open(TEST_JSONL, 'r') as f:
        lines = f.readlines()

    for line in tqdm(lines, desc="Predicting Test Set"):
        data = json.loads(line)
        img_rel_path = data.get('image_path', data.get('image', ''))
        img_path = os.path.join(IMAGE_DIR, img_rel_path)

        # 1. Safe Image Loading
        img = safe_predict_load(img_path)

        if img is not None:
            try:
                img_tensor = val_transform(img).unsqueeze(0).to(DEVICE)
                with torch.no_grad():
                    out = model(img_tensor).cpu().numpy()[0]
                    p_score = denormalize(out[0])
                    k_score = denormalize(out[1])
            except Exception as e:
                # If model forward pass fails (e.g. GPU OOM)
                p_score, k_score = 3.0, 3.0
        else:
            # 2. Handle Corrupted/Missing images with a neutral score
            # This prevents the "cannot identify image file" error from stopping the script
            p_score, k_score = 3.0, 3.0

        predictions.append({
            "id": data.get('pid', -1),
            "perception": round(float(p_score), 4),
            "knowledge": round(float(k_score), 4)
        })

    # --- Wrap and Save ---
    submission_data = {
        "team": TEAM_NAME,
        "method": MODEL_NAME,
        "track": "S",
        "predictions": predictions
    }

    with open(OUTPUT_JSON, 'w') as f:
        json.dump(submission_data, f, indent=2)

    print(f"\n✅ Done! Processed {len(predictions)} images.")
    print(f"File saved to: {OUTPUT_JSON}")
```
